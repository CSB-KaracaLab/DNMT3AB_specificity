# Pairwise flanking nucleotide (+1 to +3) - residue mapping with h-bond densities


## DNMT3A

```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
import csv
import os

sns.set(style="ticks")

input_files = os.path.join('DNMT3A_inputs')

#Please specify the DNA sequence motif here like cgc, cgg, cgt or cga
typ="cgc"

target=pd.read_csv(input_files+"/heatmap_3a_"+typ+"_target", sep=" ")
compl=pd.read_csv(input_files+"/heatmap_3a_"+typ+"_compl", sep=" ")

target['frequency'] = target['pair_freq']/3204
compl['frequency'] = compl['pair_freq']/3204

interactions_target = target.pivot("resname_a", "resid_b", "frequency")
interactions_compl = compl.pivot("resname_a", "resid_b", "frequency")

column_order_t = ["(+1)", "(+2)", "(+3)"]
column_order_c = ["(+1')", "(+2')", "(+3')"]

interactions_target_r = interactions_target.reindex(column_order_t, axis=1)
interactions_compl_r = interactions_compl.reindex(column_order_c,axis=1)

interactions_compl_r.sort_values(by=('resname_a'), ascending=True,inplace=True)


fig,(ax1,ax2) = plt.subplots(2,1
                             , figsize=(4,4.5)
                            )
g1 = sns.heatmap(interactions_target_r, ax=ax1,cmap="RdPu"
                 ,cbar_kws={'ticks': [0.0, 0.38, 0.76, 1.14, 1.52, 1.9]}
                 ,vmin=0, vmax=1.9
                 , linewidths=0.1, linecolor='black'
                 ,annot=True, annot_kws={'size': 13}
                )
g2 = sns.heatmap(interactions_compl_r,ax=ax2,cmap="RdPu"
                 ,cbar_kws={'ticks': [0.0, 0.38, 0.76, 1.14, 1.52, 1.9]}
                 ,vmin=0, vmax=1.9
                 , linewidths=0.1, linecolor='black'
                 ,annot=True, annot_kws={'size': 13}
                )

for ax in [g1,g2]:
    tl = ax.get_xticklabels()
    ax.set_xticklabels(tl, fontsize=13, fontweight='bold')
    tly = ax.get_yticklabels()
    ax.set_yticklabels(tly, rotation=60, fontsize=15, fontweight='bold')
    ax.set_xlabel("")
    ax.set_ylabel("")

    for _, spine in ax.spines.items():
        spine.set_visible(True)

fig.suptitle('DNMT3A - '+typ.upper(), fontsize=15, fontweight='bold')

plt.savefig("heatmap-3a-"+typ+"-p-annot.pdf", dpi=600, bbox_inches='tight', format="pdf")

```
