# Probability distributions of interactions during MD simulations

## Base-specific hydrogen bonds

```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
import csv
import os

input_files = os.path.join('DNMT3A_inputs')

a=pd.read_csv(input_files+"/3A-CGC-countbasespechbond", sep="\t", header=None )
a.columns = ["count"]
a['Type'] = 'CGC'

b=pd.read_csv(input_files+"/3A-CGG-countbasespechbond", sep="\t", header=None )
b.columns = ["count"]
b['Type'] = 'CGG'

c=pd.read_csv(input_files+"/3A-CGT-countbasespechbond", sep="\t", header=None )
c.columns = ["count"]
c['Type'] = 'CGT'

d=pd.read_csv(input_files+"/3A-CGA-countbasespechbond", sep="\t", header=None )
d.columns = ["count"]
d['Type'] = 'CGA'

dnmt3a_bshb = pd.concat([a, b, c, d], ignore_index=True)
dnmt3a_bshb
```

```python
import numpy as np
plt.figure(figsize=(7, 6))

palet=['#C71585','#A9A9A9', 'orange', 'lightseagreen']

sns.violinplot(x="Type", y="count", data=dnmt3a_bshb, bw=0.5, cut=0, scale='count', palette=palet)

plt.xticks(fontsize=13, fontweight='bold')
plt.yticks(np.arange(0, 19, 3))

plt.yticks(fontsize=13, fontweight='bold')
plt.ylabel('Number of h-bonds',fontsize=15, fontweight='bold')
plt.ylim(ymax = 20, ymin = -2)

plt.xlabel('',fontsize=13, fontweight='bold')
plt.title("Base-specific h-bond - DNMT3A",fontsize=20, fontweight='bold')

plt.savefig("3a-cpg-bshb-violin_label.pdf", dpi=600, bbox_inches='tight', format="pdf")
```

## Ionic bonds

```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
import csv

input_files = os.path.join('DNMT3A_inputs')

i=pd.read_csv(input_files+"/3A-CGC-countionic", sep="\t", header=None )
i.columns = ["count"]
i['Type'] = 'CGC'

j=pd.read_csv(input_files+"/3A-CGG-countionic", sep="\t", header=None )
j.columns = ["count"]
j['Type'] = 'CGG'

k=pd.read_csv(input_files+"/3A-CGT-countionic", sep="\t", header=None )
k.columns = ["count"]
k['Type'] = 'CGT'

l=pd.read_csv(input_files+"/3A-CGA-countionic", sep="\t", header=None )
l.columns = ["count"]
l['Type'] = 'CGA'


dnmt3a_ionic = pd.concat([i, j, k, l], ignore_index=True)
dnmt3a_ionic
```
